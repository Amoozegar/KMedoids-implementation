```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
import imageio
import random
import time
import sys
```

```python
def initialize_centroids(k,pixels):
    """
    Input: k:number of clusters - pixels 
    output: centroids with 3 columns and k rows 
    """
    numberOfPixels = pixels.shape[0]
    random_centroids = []
    i=0
    while (i<k):
        r = np.random.randint(numberOfPixels)
        random_centroids.append(pixels[r])
        i+=1

    centroids = np.array(random_centroids)
    return centroids
```

```python
def assign_PointsToCentroids(X,centroids):
    
    """
    Inputs: X(array of points or pixels) - centroids is the array of centroids
    Output : cclass is the array of size (255440(pixels),1)
    each element shows the index of assigned centroids to each point
    """
    numberOfPixels = X.shape[0]
    cclass = np.zeros(numberOfPixels)
    for i in range(numberOfPixels):
        temp_distance = np.zeros(len(centroids))
        for j in range(len(centroids)):
            #Manhattan distance
            temp_distance[j] = sum(abs(X[i,:] - centroids[j,:] ))
            print ('centroids', centroids)
            print ('temp_distance[j]', temp_distance[j])
            print ('X[i,:]', X[i,:])
            print ('centroids[j,:]', centroids[j,:])
            sys.exit(1)
        centroid_index = np.argmin(temp_distance)+1  
        cclass[i] = centroid_index
        
    return cclass    
```

```python
def update_centroids(pixels, cclass, centroid):
    
    """
    Input : array of pixels , class: assigned clusters to each point, centroid: array of size(k,1) consisting of centroids
    output: centroids with k rows and 3 columns
    """
    
    
    new_centroids = np.zeros((len(centroid),3))
    for j in range(len(centroid)):
        
        index_tuple = np.where(cclass==j+1)
        
        #index of all points in cluster k=j+1
        cluster_index = index_tuple[0]
        
        print 'number of points in cluster ', j+1,' : ', len(cluster_index)
        
        if (len(cluster_index) < 10):
            print ('one of clusters has zero points')

            return centroid
            break

        #randomly select 10% of this cluster to find the centroid
        sampled_index = random.sample(cluster_index, int(0.1*(len(cluster_index))))
        
        #vector to save the distance of chosen centroid to other points in that cluster
        distance_vector = np.zeros(len(sampled_index))
        
        point_index = np.zeros(len(sampled_index))
                
        cn = 0
        print 'number of sample points from cluster ',j+1 ,' : ',len(sampled_index)
        for i in sampled_index:
        
            temp_distance = 0
            temp_centroid = pixels[i]
            
            for l in cluster_index:
            
                #Manhattan distance
                temp_distance = sum(abs(pixels[l] - temp_centroid)) + temp_distance
                
            distance_vector[cn] = temp_distance
            point_index[cn] = i
            cn+=1
            if (cn%1000==0):
#                 end = time-time()
                print cn, ' points tested for cluster centroid'
                
        point_ind = np.argmin(distance_vector)

        
        if (np.count_nonzero(distance_vector ==0)>=1):
            
            zero = np.where(distance_vector==0)
            print 'zero',zero
            print 'zero distance'
            sys.exit(1)
            
        else:
            index= int(point_index[point_ind])
            
            new_centroids[j] = pixels[index]
            
        print 'cluster ',  j+1 , ' is processed'
        
    return new_centroids          
```

```python
def CalObjFunction(pixels, centroid, cclass):
    """
    calculate objective function based on the similarity function
    """
    distance = 0
    for i in range(len(cclass)):
        #Manhattan distance
        distance  = sum(abs(pixels[i] - centroid[int(cclass[i]-1)])) + distance
    
    return distance    
            
    
```
